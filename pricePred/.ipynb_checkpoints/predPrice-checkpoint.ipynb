```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn import tree
import pickle
```

```python
df = pd.read_csv("datafile.csv")
df.head(10)
```

```python
def getDataAccToCommodity(commodity):
    try:
        df1 = df[df['COMMODITIES'] == commodity]
        a = list(df1.iloc[0])
        prices = a[2:]
        commodity = a[0]
        weight = a[1]
        dates = list(df1.columns[2:])

        df2 = pd.DataFrame()
        df2['Date'] = dates
        df2['Prices'] = prices/weight
    except:
        df2 = pd.DataFrame()
    
    return df2

def getDateNum(date):
    a = date.split("-")
    month = a[0]
    year = int(a[1])
    num = year*12 + monthToNumber[month]
    return num
```

```python
crops = ['Moong']
monthToNumber = {
    "Jan":1, "Feb":2, "Mar":3, "Apr":4, "May":5, "Jun":6, "Jul":7, "Aug":8, "Sep":9, "Oct":10, "Nov":11, "Dec":12
}
```

```python
for crop in crops:
    data = getDataAccToCommodity(crop)
    data['DateModified'] = data["Date"].apply(lambda x: getDateNum(x))
    
    X = list(data["DateModified"])
    X = [[k] for k in X]
    y = data["Prices"]
    
    X = np.asarray(X)
    y = np.asarray(y)
    
    # X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
    X_train, y_train = X, y
    
    clf = tree.DecisionTreeRegressor()
    clf = clf.fit(X_train, y_train)
    
    filename = crop+'.sav'
    pickle.dump(clf, open(filename, 'wb'))
```

```python
import pickle
loaded_model = pickle.load(open('Moong.sav', 'rb'))

# inputs
year = 2021
month = 4
num = ((year-2000)*12)+month

price = loaded_model.predict([[num]])[0]
price
```
